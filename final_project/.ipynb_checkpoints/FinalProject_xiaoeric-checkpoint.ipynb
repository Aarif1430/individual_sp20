# Overview

This study aimed to reveal any correlation between the health inspection scores of facilities and the socioeconomic conditions in which they are located. We analyzed 18,466 inspection records, 189,802 food safety violations, and the income data of 38 zip codes in the state of North Carolina to possibly find such a correlation. In the analysis, we attempted to compared multiple different socioeconomic factors, including income, welfare, and ethnicity percentages to the average scores of the facilities in their zip code region. We found that there was no such correlation between any of those factors and health inspection scores.

# Research Question

Is there a relationship between the health inspection score of food facilities and the socialeconomic conditions of the region in which they are located?

## Background and Prior Work

Many studies show that there exists a correlation between food and health safety, and the socioeconomic status of a community. A study on the influence of such factors on the importance of food safety in a community finds that household meal planners who are "female, older, more educated, non-working, [or] have at-risk household members" hold the importance of food safety in high regard (Lin). In another study on urban pest prevalence, the authors conclude that bed bug infestations are more likely to occur in low income areas, due to higher rates of crowding and eviction, and that this disproportionate health safety burden is an issue of social justice (Sutherland). In order to inform public policy concerning health safety regulations, we conducted an analysis on the potential impact of socioeconomic conditions on health safety levels in the local community.

Beyond social welfare, we also find that it is also in the interest of local food businesses for us to conduct this study. A similar food safety inspection study found that chain restaurants experienced fewer violations than nonchain restaurants, and that additional inspections for nonchain restaurants helped to reduce average violations. Through this analysis, we may be able to identify local businesses in need of food safety training or unreasonably disadvantaged by their income status.

# Hypothesis


Restaurants and other food facilities located in lower-income neighborhoods are more likely to have lower health inspection scores than those in higher-income areas. # Setup

We begin by importing the necessary packages and libraries for statistical analysis later in the program, and also loading the csv files into DataFrames.

```python
# Imports and settings from A3

%matplotlib inline

# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn import linear_model
import patsy
import statsmodels.api as sm

# Configure libraries
# The seaborn library makes plots look nicer
sns.set()
sns.set_context('talk')

# Don't display too many rows/cols of DataFrames
pd.options.display.max_rows = 7
pd.options.display.max_columns = 8

# Round decimals when displaying DataFrames
pd.set_option('precision', 2)

# Read data from csv files
df_inspec = pd.read_csv("data/inspections.csv")
df_zip = pd.read_csv("data/zipcodes.csv")
df_violat = pd.read_csv("data/violations.csv")
```

Here we calculate the total number of inspections and violations for each facility and format it to be easily merged into the main DataFrame later.

```python
# Count number of inspections and violations for each facility
inspec_counts = df_inspec.groupby('hsisid').size().reset_index(name='inspection_counts').set_index('hsisid')
violat_counts = df_violat.groupby('hsisid').size().reset_index(name='violation_counts').set_index('hsisid')
```

# Data Cleaning

We clean the data to remove any unnecessary identifying information, as well as any information that is not relevant to the research question. We also aggregate all the counts from above into one DataFrame.

```python
# Clean inspection data and only keep relevant variables
df_inspec_filt = df_inspec[['hsisid','zip','score','num_critical','num_non_critical']]

# Merge inspection data with regional data by zip code
df_merged = df_inspec_filt.merge(df_zip, on='zip')

# Final aggregation of data across csv files
df_main = df_merged.set_index('hsisid').join(inspec_counts).join(violat_counts)
```

# Data Analysis & Results

Now that the data is fully aggregated, we can begin to look at the distribution of our variables. Let's look at the distribution of health inspection scores.

```python
ax = plt.subplot()
plt.hist(df_main.score)
plt.title("Distribution of health inspection score")
plt.show()
```

This distribution is incredibly skewed to the right! But if we aggregate scores by zip code and average them...

```python
# Group and aggregate data by zip code
df_main_zip = df_main.groupby('zip').mean()

ax = plt.subplot()
plt.hist(df_main_zip.score)
plt.title("Distribution of health inspection score")
plt.show()
```

The distribution is much less skewed! Aggregating scores by zip code also helps us when making scatter plots later on, because the income data is aggregated across each zip code region as well. Unfortunately, most of scores fall in this narrow range of 95-98, though the original scores didn't fall into a very large range to begin with.

Let's also look at the distribution of income data.

```python
ax = plt.subplot()
plt.hist(df_main_zip.median_family_income_dollars)
plt.title("Distribution of median family income")
plt.show()

print("Skew:",df_main_zip.median_family_income_dollars.skew())

ax = plt.subplot()
plt.hist(df_main_zip.median_household_income_dollars)
plt.title("Distribution of median household income")
plt.show()

print("Skew:",df_main_zip.median_household_income_dollars.skew())

ax = plt.subplot()
plt.hist(df_main_zip.per_capita_income_dollars)
plt.title("Distribution of per capita income")
plt.show()

print("Skew:",df_main_zip.per_capita_income_dollars.skew())
```

For the most part, these distributions aren't very skewed. Let's also look at the relationships between these three measures of income with a scatter matrix.

```python
pd.plotting.scatter_matrix(df_main_zip[['median_family_income_dollars','median_household_income_dollars','per_capita_income_dollars']], figsize=(8, 8))
```

Each of these income measurements are directly proportional to one another. This means that using either of them as the measure for income data in later scatter plots will generally exhibit the same relationships. We will demonstrate this by plotting each of these measures against health inspection score.

```python
ax = plt.subplot()
plt.scatter(df_main_zip.median_family_income_dollars, df_main_zip.score)
plt.title("Median family income vs. health inspection score")
plt.show()

print("Correlation:",df_main_zip[['median_family_income_dollars','score']].corr())

ax = plt.subplot()
plt.scatter(df_main_zip.median_household_income_dollars, df_main_zip.score)
plt.title("Median household income vs. health inspection score")
plt.show()

print("Correlation:",df_main_zip[['median_household_income_dollars','score']].corr())

ax = plt.subplot()
plt.scatter(df_main_zip.per_capita_income_dollars, df_main_zip.score)
plt.title("Per capita income vs. health inspection score")
plt.show()

print("Correlation:",df_main_zip[['per_capita_income_dollars','score']].corr())
```

We observe from the graphs themselves that there may be a slight positive correlation between income and score, however the actual calculation shows that these two variables are hardly correlated. There is no evident relationship between regional income and health inspection scores. Let's look at the other socioeconomic variables and their relationship to health inspection score, starting with poverty and welfare statistics.

```python
pd.plotting.scatter_matrix(df_main_zip[['percent_damilies_below_poverty_line','percent_snap_benefits','percent_supplemental_security_income']], figsize=(8, 8))
print("Poverty data skew:",df_main_zip.percent_damilies_below_poverty_line.skew())
print("SNAP data skew:",df_main_zip.percent_snap_benefits.skew())
print("Security income data skew:",df_main_zip.percent_supplemental_security_income.skew())
```

Each of these measurements is quite skewed! If we transform the data...

```python
pov_log = np.log(df_main_zip.percent_damilies_below_poverty_line)
snap_log = np.log(df_main_zip.percent_snap_benefits)
secur_log = np.log(df_main_zip.percent_supplemental_security_income)

ax = plt.subplot()
plt.hist(pov_log)
plt.title("Distribution of families in poverty")
plt.show()

print("Poverty data skew:",pov_log.skew())

ax = plt.subplot()
plt.hist(snap_log)
plt.title("Distribution of SNAP benefits")
plt.show()

print("SNAP data skew:",snap_log.skew())

ax = plt.subplot()
plt.hist(secur_log)
plt.title("Distribution of supplemental security income")
plt.show()

print("Security income data skew:",secur_log.skew())

```

Now the distribution is much more normal and we can proceed to calculate the correlation with health inspection score.

```python
ax = plt.subplot()
plt.scatter(pov_log, df_main_zip.score)
plt.title("Families in poverty vs. health inspection score")
plt.show()

pov_log.name = 'pov_log'
print("Correlation:",pd.DataFrame(zip(pov_log, df_main_zip.score), columns=[pov_log.name,df_main_zip.score.name]).corr())

ax = plt.subplot()
plt.scatter(snap_log, df_main_zip.score)
plt.title("SNAP benefits vs. health inspection score")
plt.show()

snap_log.name = 'snap_log'
print("Correlation:",pd.DataFrame(zip(snap_log, df_main_zip.score), columns=[snap_log.name,df_main_zip.score.name]).corr())

ax = plt.subplot()
plt.scatter(secur_log, df_main_zip.score)
plt.title("Supplemental security income vs. health inspection score")
plt.show()

secur_log.name = 'secur_log'
print("Correlation:",pd.DataFrame(zip(secur_log, df_main_zip.score), columns=[secur_log.name,df_main_zip.score.name]).corr())
```

None of these metrics are strongly correlated with health inspection score either. Our final consideration for potential socioeconomic factors is ethnicity.

```python
ax = plt.subplot()
plt.hist(df_main_zip.percent_nonwhite)
plt.title("Distribution of non-white ethnicities")
plt.show()

print("Skew:",df_main_zip.percent_nonwhite.skew())

nw_log = np.log(df_main_zip.percent_nonwhite)

ax = plt.subplot()
plt.hist(nw_log)
plt.title("Distribution of non-white ethnicities")
plt.show()

print("Skew:",nw_log.skew())
```

We normalize the distribution of non-white ethniticies and proceed to correlate it with score.

```python
ax = plt.subplot()
plt.scatter(nw_log, df_main_zip.score)
plt.title("Non-white ethnicities vs. health inspection score")
plt.show()

nw_log.name = 'nw_log'
print("Correlation:",pd.DataFrame(zip(nw_log, df_main_zip.score), columns=[nw_log.name,df_main_zip.score.name]).corr())
```

We have thus exhausted our list of potential socioeconomic factors and observe that none of them strongly correlate to health inspection scores.

# Ethics & Privacy

We recognize that the reputation of restaurants and similar business have a significant influence on their profits and it would certainly not be ethical for a data analysis to disproportionately harm or benefit business profits. As such, any information which may be used to identify a specific facility will be cleaned and not considered in the analysis. Variables that will be removed include address, name, city, telephone number, coordinates, and any other such information. Facilities will instead be grouped by their zipcode and associated with the economic data of that region.

# Conclusion & Discussion

As we cannot find a meaningful and significant correlation between health inspection and socioeconomic factors in this dataset, we fail to the reject the null hypothesis. In each scatter plot graph, we find that either there is a slight and insignificant trend or no correlation at all. As such, we also cannot produce any meaningful regression or classification model to predict health inspection scores from regional income data or vice versa.

We speculate that one major limitation is the scope of the data itself. Since all of the data was collected from one state in the United States of America, it is possible that the economic disparity represented in the dataset is not significant enough to demonstrate a correlation with food safety. It may also be possible that the state of North Carolina specifically implements equitable food safety policies to aid lower-income restaurants in order to counteract the effect that was hypothesized.

Another such limitation may also be that most of the health inspection scores were in the range of 90-100. This characteristic limited our ability to understand if a trend exists between low inspection scores and some other factor, which we had speculated to be related to either income, welfare, or ethnicity. In the future, it may be more informative to collect data about food safety inspections outside of North Carolina to compile more information about low-scoring facilities for the purpose of informing local health safety policies.
