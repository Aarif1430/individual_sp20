# Data Cleaning

The Clean_Zip function removes all zipcodes with the extension such that only the first 5 characters are used. This is used in order to merge datasets on common zipcodes.

```python
def Clean_Zip(zipcode):
    # Convert to a string
    code = str(zipcode)
    # Get only the first 5 characters
    new_zip = code[0] + code[1] + code[2] + code[3] + code[4]
    return new_zip
```

The Laplace_Rule_Of_Succession function takes in the ratings and review count from the Yelp dataset and converts the values to a new more standardized value. It does this by taking the rating out of 5 stars and converting it to a percent, then multiplying it by the number of reviews to get the number of "successes" and finally adds one to the numerator. The function then adds two to the denominator and divides the two to get the final rating rounded to 3 decimal places.

```python
def Laplaces_Rule_Of_Succession(rating, num_reviews):
    
    # Add one to the numerator
    rating = (int(rating)/5 * num_reviews) + 1
    
    # Add two to the denominator
    num_reviews = int(num_reviews) + 2
    
    # Calculate the new standardized review
    laplace_review = rating / num_reviews
    
    # Round the value to 3 digits
    return (round(laplace_review, 3))
```
