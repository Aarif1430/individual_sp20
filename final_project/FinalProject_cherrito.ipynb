# MODEL

except those values that I find out has relationship with the health inspection score for diner places. There should be other values that also has relationship with health inspection score. so I will first find the correlation of the values to the score. Then i will pick those values and put them into the model. 

For the model, I use the boosting model and also make some analyze on which parameter I should use to get a best result

```python
num_col=result2.select_dtypes(include=[np.float64,np.int])
top_corr=num_col.corr()['score'].abs().sort_values(ascending=False).drop(['score'])
top_corr= top_corr[0:9]
top_corr
```

```python
#choosing other usful variables from another dataframe i use
num_col1=new_result.select_dtypes(include=[np.float64,np.int])
top_corr1=num_col1.corr()['scoremerged_'].abs().sort_values(ascending=False).drop(['scoremerged_'])
top_corr1= top_corr1[0:10].drop(['hsisid'])
top_corr1
```

```python
column_get = top_corr.index
column_get2 = top_corr1.index
for i in column_get:
    i = i+'merged_'
    if i not in column_get2:
        column_get2.append(i)
score_index = new_result['scoremerged_']
column_get2
```

```python
#get those columns and put into the data
#as we can notice that those index are same, so we will only use more data from new_result, because it has more data, 
#which can increase the accuracy of model by using more data
modify_x=new_result[column_get2]
modify_x['score'] = new_result['scoremerged_']
modify_x = modify_x.dropna()
modify_x.reset_index(inplace = True, drop = True)
modify_x.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(modify_x, modify_x['score'], test_size=0.3, random_state=42)
```

```python
from sklearn import datasets, ensemble
from sklearn.metrics import mean_squared_error
params = {'n_estimators': 100,
          'max_depth': 15,
          'min_samples_split': 5,
          'learning_rate': 0.1,
          'loss': 'ls'}

reg = ensemble.GradientBoostingRegressor(**params)
reg.fit(X_train, y_train)

mse = mean_squared_error(y_test, reg.predict(X_test))
print('rmse: '+str(mse))
```

As you can notice from the model rmse score that it actually pretty good, which mean this model can use to predict the health inspection score of some diner places well and then used to improve the efficiency of health inspection process
