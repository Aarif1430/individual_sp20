## Prior Work 1

A Research by Darcey et al. investigated the utility of Geographic information systems (GIS) in determining whether community socioeconomic status (SES) and/or demographics relate to access to safe food, as measured by food service critical health code violations (CHV) as a proxy for risk for foodborne illness (FBI).

- Researchers accessed health inspection records documenting CHV for 10,859 food service facilities collected between 2005 and 2008 in Philadelphia, PA.
- They concluded that food service facilities in higher poverty areas had a greater number of facilities (with at least one CHV) and had more frequent inspections than facilities in lower poverty areas. The facilities in lower poverty areas, however, had a higher average number of CHV per inspection. 
- They also found out that Hispanic facilities had more CHV than other demographics, and Hispanic and African American facilities had fewer days between inspections

Here we checked the phone numbers, and noticed that some of them have external lines, so we created a new column called external specifically, and all non-nan phone numbers were successfully transformed to numerical phone numbers

```python
# Convert phone number into actual numbers
# create a new column called phone external
phones = list(inspDF["phone"])
ext = []
notNum = []
for i in range(len(phones)):
    if str(phones[i]) == "nan":
        ext.append(np.nan)
        continue
    curr = phones[i]
    curr = curr.replace("(","")
    curr = curr.replace(")","")
    curr = curr.replace("-","")
    curr = curr.replace(" ","")
    curr = curr.split("ext.")
    if len(curr) == 2:
        ext.append(curr[1])
    else:
        ext.append(np.nan)
    
    phones[i] = curr[0]
    
    try:
        phones[i] = int(phones[i])
    except ValueError:
        notNum.append(i)
inspDF["phone"] = phones
inspDF["external"] = ext
inspDF
```

We know that there may be restaurants sharing same longitude and latitude, thus we use phone numbers for supplementary. As we see that only 193 cases of restaurant phone is missing, which would almost ensure that we would refer to the corresponding restaurant while analyzing data from yelp.

```python
def checkRange(inspDF):
    # Trim inspection date
    dates = list(inspDF["date"])
    for i in range(len(dates)):
        dates[i] = dates[i][0:10]
    inspDF["date"] = dates

    #Check the longitude(-180~180),latitude(-180~180),zip(27501~27703), inspN(1~50),score(0~100)
    #, and num_critical(0~50)

    days = list(inspDF["openDays"])
    long = list(inspDF["longitude"])
    lat = list(inspDF["latitude"])
    zipcode = list(inspDF["zip"])
    inspN = list(inspDF["inspN"])
    score = list(inspDF["score"])
    numC = list(inspDF["numCritical"])
    wrongZips = []


    for i in range(len(inspDF)):
        if str(days[i]) != "nan":
            if days[i] < 0 or days[i] > 10000:
                print("day of row " + str(i)+" out of range")
        if str(long[i]) != "nan":
            if long[i] < -180 or long[i] >180:
                print("longitude of row " + str(i) + " out of range")
        if str(lat[i]) != "nan":
            if lat[i] < -180 or lat[i] >180:
                print("latitude of row " + str(i) + " out of range")
        if str(zipcode[i]) != "nan":    
            if zipcode[i] < 27501 or zipcode[i] >27703:
                print("zipcode of row " + str(i) + " out of range")
                print(zipcode[i])
                wrongZips.append(i)

        if str(inspN[i]) != "nan":
            if inspN[i] < 0 or inspN[i] >50:
                print("Inspection number of row " + str(i) + " out of range")
        if str(score[i]) != "nan":
            if score[i] < 0 or score[i] >100:
                print("Score of row " + str(i) + " out of range")
        if str(numC[i]) != "nan":
            if numC[i] < 0 or numC[i] >50:
                print("Critical number of row " + str(i) + " out of range")
                
    return wrongZips
```

## Analysis

First I want to perform Principle Components Analysis, to reduce the dimension of some features, and to see how they cluster

```python
# separate our dependent variables, score and numCritical are the results we want to predict later on
aDF = inspDF.loc[:,['openDays','family', 'houseHold', 'capita', 'rating','review', 'price']]
aDF = aDF.dropna()
inspDF = inspDF.loc[aDF.index,:]
aDF
```

```python
pca =PCA()
pca.fit(aDF)
```

```python
# PCs
print(pca.components_)
```

```python
# These are the variance explained by each PC
print(pca.explained_variance_ratio_)
```

```python
var_exp = pd.DataFrame(pca.explained_variance_ratio_,
                       ['PC1', 'PC2','PC3','PC4','PC5','PC6','PC7'])
var_exp.plot.bar(rot=0)
plt.ylabel('variance explained');
```

From this spree graph, we can see that our first principle component explains 92.12% of the variances, and the second explains 5.69% of variances, the third explains 1.58%, we can safely drop the rest of PCs without losing too much information.

Together we got 99.39% of total information from PC1, PC2, and PC3

```python
# transform the original data to PC data
res_PCs = pca.fit_transform(aDF)
pca_out = pd.DataFrame(res_PCs, 
                       columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7'])
pca_out['score'] = inspDF['score']
pca_out['numCritical'] = inspDF["numCritical"]
```

```python
# Drop all nan score/ numCriticals
pca_out = pca_out.dropna()

# only pick PC1,PC2,PC3
pca_out = pca_out.loc[:,["PC1","PC2","PC3","score","numCritical"]]
pca_out
```

```python
# Create ranges as categorical variables so as to visualize PC clusters

scoreRange = []
for i in list(pca_out["score"]):
    if i > 98:
        scoreRange.append("> 98")
    elif i >96:
        scoreRange.append("> 96")
    elif i >94:
        scoreRange.append("> 94")
    elif i >92:
        scoreRange.append("> 92")
    elif i >90:
        scoreRange.append("> 90")
    else:
        scoreRange.append("<= 90")
pca_out["scoreRange"] = scoreRange



numRange = []
for i in list(pca_out["numCritical"]):
    if i  == 0:
        numRange.append("0")
    elif i >= 1 and i <= 3:
        numRange.append("1 ~ 3")
    elif i >=4 and i <= 6:
        numRange.append("4 ~ 6")
    elif i >= 7 and i <=9:
        numRange.append("7 ~ 9")
    else:
        numRange.append(">= 10")
pca_out["critRange"] = numRange
```
